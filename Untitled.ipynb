```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("star.csv")
df.head()
```

```python
df.info()
```

```python
pd.factorize(df["Star color"].tolist())[1]
```

```python
for i, row in df.iterrows():
    string = ''.join(e for e in row["Star color"].lower() if e.isalnum())
    if "yellow" in string and "white" in string:
        string = "yellowwhite"
    if "yellowish" in string:
        string = "yellow"
    if "whitish" in string:
        string = "white"
    df.loc[i, "Star color"] = string

df.head()
```

```python
pd.factorize(df["Star color"].tolist())[1]
```

```python
pd.factorize(df["Spectral Class"].tolist())[1]
```

```python
df.sample(n=1)
```

```python
df["Star color"] = pd.factorize(df["Star color"].tolist())[0]
df["Spectral Class"] = pd.factorize(df["Spectral Class"].tolist())[0]

df.head()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X = df.drop("Star type", axis=1)
Y = df["Star type"].values

X.head()
```

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel='rbf')
clf.fit(X, Y)
```

```python
y_pred = clf.predict(X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(y_pred, Y)
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
y_pred = clf.predict(X)
y_pred
```

```python
tree.plot_tree(clf)
```

```python
accuracy_score(Y, y_pred, normalize=False)
```
